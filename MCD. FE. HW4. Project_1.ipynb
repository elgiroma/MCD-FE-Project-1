# Downloading data

```python
# Importing dataframes
import os  # Para manejo de archivos y directorios
import urllib.request # Una forma estandard de descargar datos
import datetime # Fecha de descarga
import pandas as pd # Solo para ver el archivo descargado
import zipfile # Descompresión de archivos
```

```python
# pwd
print(os.getcwd())

#  Estos son los datos que vamos a descargar y donde vamos a guardarlos

# Datos de casos de covid en México el día 18 de agosto de 2022
COVID_url_1 = "https://datosabiertos.salud.gob.mx/gobmx/salud/datos_abiertos/historicos/2022/08/datos_abiertos_covid19_18.08.2022.zip"
COVID_archivo_1 = "COVID19_MEX_18_08_22.zip"
subdir_1 = "./data/COVID19_MEX_18_08_22_"
info_1 = """
        18 de agosto de 2022
        Recolección de información estadística de los casos del virus COVID 19 presentados en México,
        contiene diferentes características sobre los pacientes como son el sexo, municipio, fecha de 
        sintomas, edad, embarazo, diabetes, obesidad, entre muchas otras.
        """ 

# Datos de casos de covid en México el día 18 de septiembre de 2022
COVID_url_2 = "https://datosabiertos.salud.gob.mx/gobmx/salud/datos_abiertos/historicos/2022/09/datos_abiertos_covid19_18.09.2022.zip"
COVID_archivo_2 = "COVID19_MEX_18_09_22.zip"
subdir_2 = "./data/COVID19_MEX_18_09_22_"
info_2 = """
        18 de septiembre de 2022
        Recolección de información estadística de los casos del virus COVID 19 presentados en México,
        contiene diferentes características sobre los pacientes como son el sexo, municipio, fecha de 
        sintomas, edad, embarazo, diabetes, obesidad, entre muchas otras.
        """ 

# Datos de casos de covid en México el día 18 de octubre de 2022
COVID_url_3 = "https://datosabiertos.salud.gob.mx/gobmx/salud/datos_abiertos/datos_abiertos_covid19.zip"
COVID_archivo_3 = "COVID19_MEX_18_10_22.zip"
subdir_3 = "./data/COVID19_MEX_18_10_22_"
info_3 = """
        18 de octubre de 2022
        Recolección de información estadística de los casos del virus COVID 19 presentados en México,
        contiene diferentes características sobre los pacientes como son el sexo, municipio, fecha de 
        sintomas, edad, embarazo, diabetes, obesidad, entre muchas otras.
        """ 
```

```python
# Function to download the datasets
def downloading_dataset(dataset_url, dataset_name, dataset_subdir, info_text):
    if not os.path.exists(dataset_name):
        if not os.path.exists(dataset_subdir):
            os.makedirs(dataset_subdir)
        urllib.request.urlretrieve(dataset_url, dataset_subdir + dataset_name)  
        with zipfile.ZipFile(dataset_subdir + dataset_name, "r") as zip_ref:
            zip_ref.extractall(dataset_subdir)

        with open(dataset_subdir + "info.txt", 'w') as f:
            f.write("Información estadística de los casos de COVID19 en México\n18 de agosto de 2022")
            info = info_text
            f.write(info + '\n')
            f.write("Descargado el " + datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S") + "\n")
            f.write("Desde: " + dataset_url + "\n")
            f.write("Nombre: " + dataset_name + "\n")
```

```python
# Downloading the 3 datasets
downloading_dataset(COVID_url_1, COVID_archivo_1, subdir_1, info_1)
downloading_dataset(COVID_url_2, COVID_archivo_2, subdir_2, info_2)
downloading_dataset(COVID_url_3, COVID_archivo_3, subdir_3, info_3)
```

```python
# Reading dataframes
df_1 = pd.read_csv('./data/COVID19_MEX_18_08_22_/220818COVID19MEXICO.csv')
df_2 = pd.read_csv('./data/COVID19_MEX_18_09_22_/220918COVID19MEXICO.csv')
df_3 = pd.read_csv('./data/COVID19_MEX_18_10_22_/221018COVID19MEXICO.csv')
```

# Generating tidy dataframes

```python
# Concatenating all dataframes into one
df = pd.concat([df_1, df_2, df_3], axis=0)

# Dropping these columns because we won't be using them
df = df.drop(columns=['FECHA_ACTUALIZACION', 'ID_REGISTRO', 'PAIS_NACIONALIDAD','PAIS_ORIGEN'])

# Resetting index in dataframe
df = df.reset_index(drop=True)
```

```python
# Changing dates str type to datetime type
df['FECHA_INGRESO'] = pd.to_datetime(df['FECHA_INGRESO'])
df['FECHA_SINTOMAS'] = pd.to_datetime(df['FECHA_SINTOMAS'])

# 9999-99-99 is out of bounds for to_datetime so i set it to the maximum date possible
df.loc[(df['FECHA_DEF'] == '9999-99-99'),'FECHA_DEF']= '2262-04-11'
df['FECHA_DEF'] = pd.to_datetime(df['FECHA_DEF'])
```
